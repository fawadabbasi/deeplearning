```r
# Load the library into R workspace.
library("xlsx")
```

```r
data <- read.xlsx("dataama.xlsx", sheetIndex = 1)
print(data)
```

```r
input <-lm(vc..m.min. ~ f..mm.+Bahn+Fc..N.+Stdabw..Fc..N.,data)
summary(input) # show results
#print(head(input))
```

```r
coefficients(input) # model coefficients
```

```r
confint(input, level=0.95) # CIs for model parameters 
```

```r
fitted(input) # predicted values
```

```r
residuals(input) # residuals
```

```r
anova(input) # anova table 
```

```r
vcov(input) # covariance matrix for model parameters 
```

```r
influence(input) # regression diagnostics
```

```r
# diagnostic plots 
layout(matrix(c(1,2,3,4),2,2)) # optional 4 graphs/page 
plot(input)
```

```r
# compare models
input1 <-lm(vc..m.min. ~ f..mm.+Fc..N.+Stdabw..Fc..N.,data)
input2 <-lm(vc..m.min. ~ f..mm.+Fc..N.,data)
anova(input1, input2)
#input1 <- lm(y ~ x1 + x2 + x3 + x4, data=mydata)
```
